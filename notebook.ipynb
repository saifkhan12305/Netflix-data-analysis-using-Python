# Data Cleaning

```python
df.head()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df.head()
```

```python
# drop the columns

df.drop(['Overview', 'Original_Language', 'Poster_Url'], axis=1, inplace=True)
df.head()
```

Converting the values of "**Vote_Average"** field into four **Categories (Popular, Average, Below_Average, Not Popular)**

```python
def categorize_col(df, col, labels):

    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']
             ]
    
    df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
```

```python
labels = ['not_popular', 'below_average', 'average', 'popular']

categorize_col(df, 'Vote_Average', labels)

df['Vote_Average']
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
# Dropping the duplicates or NaN values
df.dropna(inplace=True)

# Checking
df.isna().sum()
```

```python
df.head()
```

Splitting the **Genre** column in to a list and then explode the dataframe to have only one **Genre** per row for each movie

```python
# Splitting into a list
df['Genre'] = df['Genre'].str.split(", ")

# Explode the list
df = df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
# Casting Column into category
df['Genre'] = df['Genre'].astype('category')

df['Genre'].dtype
```

# Data Visualization

Here we'd use matplotlib and seaborn for making some informative visual and gain insight of our data

```python
# setting up seaborn configuration
sns.set_style('darkgrid')
```

#### Q.1  What is the most frequent genre of movies released on Netflix?

```python
df['Genre'].describe()
```

```python
sns.catplot(y='Genre', data= df, kind= 'count', order= df['Genre'].value_counts().index)
plt.title("Genre Distribution")
plt.show()
```

Here we see that the Genre **Drama** is most frequent Genre in the dataset
